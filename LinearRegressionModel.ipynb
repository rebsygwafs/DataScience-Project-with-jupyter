```python
                           INTRODUCTION

#Linear Regression Analysis to determine the dependency of  "Total vehicles ownership (millions)" on
#"Population (millions)".This will help identify the significant relationship  between the dependent
#and independent variables. where y = Population  (millions) and x = Total vehicles(millions)
```

```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.formula.api as smf
```

```python
#importing the dataset
dt = pd.read_csv("C:/Users/user/Downloads/Project (3)/Project/Cars_dataset.csv", header=0, sep =",")
```

```python
dt.head(20)
```

```python
# defining the variables
x = dt["Population (millions)"]
y = dt["Total vechicles (millions)"]
```

```python
# Getting important key values
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
#definning a function to return new value
def funct(X):
    return slope*X + intercept
```

```python
# plotting  a scattered graph
dt.plot(x="Population (millions)",y="Total vechicles (millions)",kind="scatter")
plt.plot(x,slope*x+intercept)
plt.ylim(ymin=0, ymax=60)
plt.xlim(xmin=0, xmax=60)
plt.xlabel ("Population (millions)")
plt.ylabel ("Total vechicles (millions)")        
plt.show()
```

```python
# the scattered graph shows there is a strong positive  relationship between the dependent and
# independent variables
```

```python
## fitting a model
model= smf.ols('y~x', data = dt)
```

```python
# calling  the summary to get the linear regression result
results = model.fit()
print(results.summary())
```

```python
# the output of the linear equation 
#y = - 0.9385 + 0.5804(Population) #which can be written as
#y = 0.5804(Population) - 0.9385
# this means that for every increase in unit population,the total vehicle ownership increases by
#0.5809
```

```python
# defining a function for predictions
def predict_veh(Population):
    return(0.5804*Population-0.9385)
print(predict_veh(67))
print(predict_veh(80))
```

```python
# the p-value of 0.000 which is a very very low p-value (mostly this is always an approximation of a 
#very low decimal number with a lot of zeros after the decimal point). this number indicates that the
# population is significant over total vehicle ownership.
# the adjusted r-squared value which is 0.973 shows that 97.3% of total variation of the total vehicle
#ownership can be explained by the population.

```
